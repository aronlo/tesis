# Lectura de archivos

```python
tempData = []

for root, dirs, files in os.walk("./data/output_numpy/passphrases", topdown=False):
    tempObj = {}
    
    vector = []
    for name in files:
      #print(os.path.join(root, name))
      f = open(os.path.join(root, name), 'rt') 
      if name == "l_rr.txt":
          val = f.read().split()
          tempObj["l_rr"] = val
          vector.extend(val)
      elif name == "l_total.txt":
          tempObj["l_total"] = f.read().split()
      elif name == "password.txt":
          tempObj["password"] = f.read()
      elif name == "p_pp.txt":
          val = f.read().split()
          tempObj["p_pp"] = val
          vector.extend(val)
      elif name == "p_pr.txt":
          val =  f.read().split()
          tempObj["p_pr"] = val
          vector.extend(val)

      elif name == "p_release_codes.txt":
          tempObj["p_release_codes"] = f.read().split()
      elif name == "p_rp.txt":
          val = f.read().split()
          tempObj["p_rp"] = val
          vector.extend(val)
      elif name == "p_rr.txt":
          val = f.read().split()
          tempObj["p_rr"] = val
          vector.extend(val)
      elif name == "p_total.txt":
          tempObj["p_total"] = f.read()
      elif name == "userid.txt":
          tempObj["userid"] = f.read()
          #print(tempObj)
          tempObj["vector"] = vector
          if len(vector) == 86 or len(vector) ==82:
              tempData.append(tempObj)
          tempObj = {}
      elif name == "date.txt":
          tempObj["date"] = f.read()
      elif name == "genuine.txt":
          tempObj["genuine"] = f.read()
      elif name == "login.txt":
          tempObj["login"] = f.read()
          
      elif name == "l_pp.txt":
          val = f.read().split()
          tempObj["l_pp"] = val
          vector.extend(val)
      elif name == "l_pr.txt":
          val = f.read().split()
          tempObj["l_pr"] = val
          vector.extend(val)
      elif name == "l_release_codes.txt":
          tempObj["l_release_codes"] = f.read().split()
    
      elif name == "l_rp.txt":
          val = f.read().split()
          tempObj["l_rp"] = val
          vector.extend(val)
      f.close()

df = pd.DataFrame(tempData)
```
